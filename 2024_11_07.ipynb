```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# 產生隨機數據 DataFrame
df2 = pd.DataFrame(np.random.randn(100, 4), columns=list("ABCD"))

# 定義 y 和 X
y = df2['B']
X = df2.iloc[:, [0, 2]]  # 選取 A 和 C 作為特徵

# 加入截距項 (常數)
X = sm.add_constant(X)

# 建立複迴歸模型
model = sm.OLS(y, X).fit()

# 顯示迴歸結果摘要
print(model.summary())
```
