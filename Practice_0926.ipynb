```python
def tf(value):
    return 2 * value - 1

test_value = [0.0, 0.0011, 0.25, 0.5, 0.54, 0.75, 1.0]
scaled_values = [tf(val) for val in test_value]  # Apply tf function to each element in test_value

print("Original values:", test_value)
print("Scaled values:", scaled_values)
```

```python
def a(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n + a(n-1)

print(a(3))
```
