```python
n = 5  # 欲生成的資料數量
x = np.random.randint(1, 10, size=n)
y = np.random.randint(1, 10, size=n)
z = np.zeros(n)  # 起始 z 座標設為 0
dz = np.random.randint(1, 5, size=n)  # 隨機生成每個柱狀的高度

# 將資料存入 DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'z': z, 'dz': dz})
colors = plt.cm.viridis(np.linspace(0, 1, n))  # 使用 viridis 顏色映射來生成 n 個顏色

# 繪製 3D 柱狀圖
fig = plt.figure(figsize=(6, 6))
ax = plt.subplot(projection='3d')

# 使用 DataFrame 中的資料繪製
for i in range(n):
    ax.bar3d(df['x'][i], df['y'][i], df['z'][i], dx=1, dy=1, dz=df['dz'][i], color=colors[i])
# 設定標題和軸標籤
ax.set_title('Randomly Generated 3D Bar Chart')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# 顯示圖表
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# 隨機生成資料
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))

#長條圖
df['Month'] = df.index.month

#繪製每個月份各欄位的平均值
df_melted = df.melt(id_vars='Month', value_vars=['A', 'B', 'C', 'D'])
sns.barplot(data=df_melted, x='Month', y='value', hue='variable')
plt.title('Average Value by Month')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.legend(title='Columns')
plt.show()

# 繪製堆疊長條圖
df.plot(kind='bar', stacked=True, figsize=(14, 6))
plt.title('Stacked Bar Chart of All Columns')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(title='Columns')
plt.show()


#圓餅圖
# 新增月份欄位
df['Month'] = df.index.month

# 按月份分組，計算各欄位的總和
monthly_sums = df.groupby('Month').sum().sum(axis=1)

# 繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(monthly_sums, labels=monthly_sums.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Total Sum by Month')
plt.axis('equal')
plt.show()

#3D柱狀長條圖
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 設定 X, Y, Z 軸的資料
x_pos = np.arange(len(df.index))
y_pos = np.arange(len(df.columns))
x_pos, y_pos = np.meshgrid(x_pos, y_pos, indexing='ij')
x_pos = x_pos.flatten()
y_pos = y_pos.flatten()
z_pos = np.zeros_like(x_pos)

# 柱狀圖的高度（數值）
values = df.values.flatten()

# 設定柱狀圖的寬度與深度
dx = dy = 0.5
dz = values

# 繪製 3D 柱狀圖
ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, shade=True)

# 設定軸標籤
ax.set_xlabel('Date')
ax.set_ylabel('Columns')
ax.set_zlabel('Values')
ax.set_xticks(np.arange(0, len(df.index), step=10))
ax.set_xticklabels(df.index[::10].strftime('%Y-%m-%d'), rotation=45)
ax.set_yticks(np.arange(len(df.columns)))
ax.set_yticklabels(df.columns)

plt.title('3D Bar Chart of DataFrame')
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# 隨機生成三個班級的成績
np.random.seed(0)  # 設定隨機種子，確保結果可重現

# 隨機生成每個班級的成績
group1 = np.random.randint(60, 100, size=10)  # 生成10個在60到100之間的隨機整數
group2 = np.random.randint(60, 100, size=12)  # 生成12個隨機整數
group3 = np.random.randint(60, 100, size=8)   # 生成8個隨機整數

# 將資料轉換為 DataFrame 格式
data = {
    'Group1': group1,
    'Group2': group2,
    'Group3': group3
}

df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data.items()]))

# 使用 scipy.stats.f_oneway 進行單因子 ANOVA 檢定
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# 輸出結果
print("F 統計量:", f_statistic)
print("p-value:", p_value)

# 解釋結果
alpha = 0.05  # 顯著性水平
if p_value < alpha:
    print("拒絕原假設：不同組別之間的均值存在顯著差異。")
else:
    print("未能拒絠原假設：不同組別之間的均值沒有顯著差異。")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
# 創建數據集
data = {
'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81,
79, 84, 77, 80],
'Method': ['A', 'A', 'A', 'A', 'A', 'A', # 教學方法 A
'B', 'B', 'B', 'B', 'B', 'B', # 教學方法 B
'C', 'C', 'C', 'C', 'C', 'C'] # 教學方法 C
}
# 將數據轉換為 DataFrame
df = pd.DataFrame(data)
# 使用 statsmodels 中的 ols 方法進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
# 輸出 ANOVA 表
print(anova_table)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np

# 隨機生成來自不同教學方法的分數數據
np.random.seed(0)

# 隨機生成 4 種方法（A, B, C, E）的分數
method_a_scores = np.random.randint(70, 100, size=6)  # 方法 A 的 6 個學生
method_b_scores = np.random.randint(70, 100, size=6)  # 方法 B 的 6 個學生
method_c_scores = np.random.randint(70, 100, size=6)  # 方法 C 的 6 個學生
method_e_scores = np.random.randint(70, 100, size=6)  # 方法 E 的 6 個學生

# 將數據組織為字典
data = {
    'Score': np.concatenate([method_a_scores, method_b_scores, method_c_scores, method_e_scores]),
    'Method': ['A'] * 6 + ['B'] * 6 + ['C'] * 6 + ['E'] * 6  # 四種方法
}

# 將數據轉換為 DataFrame
df = pd.DataFrame(data)

# 使用 statsmodels 中的 ols 方法進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# 輸出 ANOVA 表
print(anova_table)
```
